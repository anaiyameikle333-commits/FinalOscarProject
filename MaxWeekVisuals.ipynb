```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics
import math
```

```python
df = pd.read_csv('box_office_summary.csv') #importing from csv that contains maximum weeks information
```

```python
# Getting rid of outliers (Maximum weeks >= 500)
df = df[df["Rank"] <= 500]
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Rank"], bins=15, edgecolor="black")

plt.xlabel("Max Weeks in Box Office")
plt.ylabel("Count")
plt.title("Distribution of Box Office Run Lengths from the Past (1991 - 1997) and Current (2015 - 2021) Era")

plt.tight_layout()
plt.show()
```

# Past Era

```python
past_era_weeks = df.head(94)

plt.figure(figsize=(8,5))
plt.hist(past_era_weeks["Rank"], bins=20, edgecolor="black")

plt.xlabel("Max Weeks in Box Office")
plt.ylabel("Number of Movies")
plt.title("Past Era (1991 - 1997) vs. Weeks")
plt.tight_layout()
plt.show()
```

```python
past_era_weeks = df.head(94)["Rank"]
```

```python
print(statistics.mean(past_era_weeks)) # Example: calculate mean
print(statistics.median(past_era_weeks))
print(statistics.mode(past_era_weeks))
```

```python
old_era_mean = statistics.mean(past_era_weeks) # Example: calculate mean
old_era_median = statistics.median(past_era_weeks)
old_era_mode = statistics.mode(past_era_weeks)
print(old_era_mean)
print(old_era_median)
print(old_era_mode)
```

```python
old_era_mean = round(old_era_mean)
old_era_median = math.ceil(old_era_median) #original median is 22.5, I want to round up so used math.ceil 
old_era_mode = round(old_era_mode)
print(f"The mean is {old_era_mean}.")
print(f"The median is {old_era_median}.")
print(f"The mode is {old_era_mode}.")
```

# Current Era

```python
current_era_weeks = df.tail(91)

plt.figure(figsize=(8,5))
plt.hist(current_era_weeks["Rank"], bins=20, edgecolor="black")

plt.xlabel("Max Weeks in Box Office")
plt.ylabel("Number of Movies")
plt.title("Current Era (2015 - 2021) vs. Weeks")
plt.tight_layout()
plt.show()
```

```python
current_era_weeks = df.tail(91)["Rank"]
```

```python
current_era_mean = statistics.mean(current_era_weeks) 
current_era_median = statistics.median(current_era_weeks)
current_era_mode = statistics.mode(current_era_weeks)
```

```python
print(current_era_mean)
print(current_era_median)
print(current_era_mode)
```

```python
current_era_mean = round(current_era_mean)
current_era_median = round(current_era_median)
current_era_mode = round(current_era_mode)
print(f"The mean is {current_era_mean}.")
print(f"The median is {current_era_median}.")
print(f"The mode is {current_era_mode}.")
```
