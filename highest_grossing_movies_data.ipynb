```python
import pandas as pd
import requests
import time
from io import StringIO

# Define a user agent to mimic a browser
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

# years to analyze
years = list(range(1991, 1998)) + list(range(2015, 2022))

all_years_data = []

for year in years:
    print(f"Scraping top grossing movies for {year}...")
    top_gross_url = f"https://www.the-numbers.com/market/{year}/top-grossing-movies"
    
    # respect crawl delay
    time.sleep(5)
    
    response = requests.get(top_gross_url, headers=headers)
    
    if response.status_code == 200:
        # read the first table on the page
        tables = pd.read_html(StringIO(response.text), header=0)
        df = tables[0]
        
        # add year column
        df["Year"] = year
        
        all_years_data.append(df)
        
    else:
        print(f"Failed for {year}: Status {response.status_code}")

# combine all years into one dataframe
top_grossing_df = pd.concat(all_years_data, ignore_index=True)



```

```python
# save to CSV
top_grossing_df.to_csv("top_grossing_movies.csv", index=False)

print("Saved to csv")
```
